## Write a Python program to compute following computation on matrix:
a) Addition of two matrices<br>
b) Subtraction of two matrices<br>
c) Transpose of a matrix<br> 
d) Lower Triangular of a matrix<br>
e) Upper Triangular of a matrix <br>
f) Principal Diagonal of a matrices<br>
g) Secondary Diagonal of a matrices<br>
h) Multiplication of two matrices

```python
def matrix_add(m1,r1,c1,m2,r2,c2):
    if (r1==r2 and c1==c2):
        final=[None]*r1
        for i in range(r1):
            final[i]=[None]*c1
            for j in range (c1):
                final[i][j]=m1[i][j]+m2[i][j]
        print ("The Addition of the Matrix Provided  ")
        display(final,r1,c1)
    else :
        print ("Incorrect Dimension Entered ! Please try Again\n")
```

```python
def matrix_sub(m1,rm1,cm1,m2,rm2,cm2):
    if (rm1==rm2 and cm1==cm2):
        final=[None]*rm1
        for i in range(rm1):
            final[i]=[None]*cm1
            for j in range (cm1):
                final[i][j]=m1[i][j]-m2[i][j]
        print ("The Subtraction of the Matrix Provided   ")
        display(final,rm1,cm1)
    else :
        print ("Incorrect Dimension Entered ! Please try Again\n")
```

```python
def matrix_transpose(m1,rm1,cm1):
    final=[None]*cm1
    for i in range(rm1):
        final[i]=[None]*rm1
        for j in range (cm1):
            final[i][j]=m1[j][i]
    print ("The Transpose of the Matrix Provided   ")
    display(final,cm1,rm1)
```

```python
def matrix_lowertriangle(m,r,c):
    flag=0
    if (r==c):
        for i in range(r):
            for j in range (i+1,c):
                if (m[i][j]!=0):
                    flag=1
                    break
        if (flag==0):
            print ("Yeah !! The Matrix Provided is Lower Triangular ")
            display(m,r,c)
        else :
            print ("Opps ! The Matrix Provided is not Lower Triangular ")
            display(m,r,c)
    else :
        print ("Incorrect Dimension Entered ! Please try Again\n")
```

```python
def matrix_uppertriangle(m,r,c):
    flag=0
    if (r==c):
        for i in range(1,r):
            for j in range (0,i):
                if (m[i][j]!=0):
                    flag=1
                    break
        if (flag==0):
            print ("Yeah!! The Matrix Provided is Upper Triangular ")
            display(m,r,c)
        else :
            print ("Opps ! The Matrix Provided is not Upper Triangular ")
            display(m,r,c)
    else :
        print ("Incorrect Dimension Entered ! Please try Again\n")
```

```python
def matrix_pdiagonal(m,r,c):
    print("The Principal Diagonal Elements of The Provided Matix ")
    for i in range(r):
        print (m[i][i])                                 #or u can apply  i==j using if condition
    print("\n================================================================================================================\n")
```

```python
def matrix_sdiagonal(m,r,c):
    print("The Secondary Diagonal Elements of The Provided Matix ")
    for i in range(r):
        print (m[i][r-i-1])                                        #or u can apply  i+j==n-1 using if condition
    print("\n================================================================================================================\n")
```

```python
def matrix_multi(m1,r1,c1,m2,r2,c2):
    if (r2 ==c1):
        final=[None]*r1
        for i in range(r1):
            final[i]=[None]*c2
            for j in range (c2):
                Temp=0
                for k in range(0,c1):
                    Temp += (m1[i][k]*m2[k][j])
                    final[i][j]=Temp
        print ("The MUltiplication of the Matrix Provided  ")
        display(final,r1,c2)
    else :
        print ("Incorrect Dimension Entered ! Please try Again\n")
```

```python
def display(m,r,c):
    for i in range(0,r):
        for j in range (0,c):
            print (m[i][j],end="  ")
        print(" ")
    print("\n================================================================================================================\n")
```

```python
def get_matrix():
    r=int (input ("Enter The Number Of Rows for a Matrix : "))
    c=int (input ("Enter The Number Of Columns for a  Matrix : "))
    final=[None]*r
    for i in range(r):
        final[i]=[None]*c
        for j in range (c):
            n=int(input("Enter The Number For Row [{}] and Column [{}] : ".format(i,j)))
            final[i][j]=n
    return final,r,c
```

```python
def menu_call():
    print("-----------------------Welcome to Python program to compute following computation on matrix-----------------------\n"
    "\n"
    "\t \t \t 1. Addition of two matrices\n"
    "\t \t \t 2. Subtraction of two matrices\n"
    "\t \t \t 3. Transpose of a matrix\n"
    "\t \t \t 4. Lower Triangular of a matrix\n"
    "\t \t \t 5. Upper Triangular of a matrix\n"
    "\t \t \t 6. Principal Diagonal of a matrix\n"
    "\t \t \t 7. Secondary Diagonal of a matrix\n"
    "\t \t \t 8. Multiplication of two matries\n"
    "\t \t \t 9. Exit This Program . ")
    print("\n================================================================================================================\n")
    ch=int(input("Enter The Serial Number Of The Operation You Want To Perform : "))
    return ch
```
